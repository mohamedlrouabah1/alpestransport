```python
path= root_path+'referentiel-gares-voyageurs.csv'
gares_df = pd.read_csv(path, sep=';')
```

```python
gares_df['Code postal'].astype(int)
#gares_df['Code Commune'].astype(int)
#garesdf['Code département'].astype(int)
```

On ne garde que les gares de la région **Auvergne Rhône-Alpes**:

```python
gare_region_df = pd.concat([gares_df[gares_df['Région SNCF'] == 'REGION AUVERGNE'], gares_df[gares_df['Région SNCF'] == 'REGION ALPES']])
gare_region_df = pd.concat([gare_region_df, gares_df[gares_df['Région SNCF'] == 'REGION RHONE-ALPES']])
```

```python
gare_region_df.head().transpose()
```

On supprime les colonnes dont on n'aurra pas besoin

```python
try:
    del gare_region_df['Date fin validité plateforme']
    del gare_region_df['Segment DRG']
    del gare_region_df['Niveau de service']
    del gare_region_df['RG']
    del gare_region_df['TVSs']
    del gare_region_df['SOPs']
    del gare_region_df['Gare']
    del gare_region_df['Intitulé fronton de gare'] # duplication de la colonne 'Intitulé gare'
    del gare_region_df['Gare DRG']
    del gare_region_df['Gare étrangère']
    del gare_region_df['DTG']
    del gare_region_df['Unité gare'] # idem que colonne 'Région SNCF'
    del gare_region_df['UT'] # duplication de 'Intitulé plateforme' en upper case
    del gare_region_df['Nbre plateformes'] # tjrs 1 donc pas utile de la conserver
    del gare_region_df['TVS']
    del gare_region_df['WGS 84'] # duplication des colonnes 'Longitude' et 'Lattitude', à commenter si l'on préfère utiliser ce format de données pour les coords
except:
    pass
```
