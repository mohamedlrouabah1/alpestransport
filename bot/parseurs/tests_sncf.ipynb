Nous considérons les 2 formats intermédiaires suivants:
- pandas dataframe : avec 1 ligne = une entité, et une colonne = une propriété.
- un dictionnaire python dont le format est donné ci-dessous.

Format du dictionnaire python qui peut être utilisé pour transformer les données avant insertion:
```py
my_new_item = {
    'entity':'item',
    'label': 'my_great_label'
    'description': 'just in case...'
    'property': [
        {
            'entity':'property',
            'type': 'string', # item, property, quantity, lexeme,  Geographic coordinates, Date, Time
            'label':"instance of",
            'value':'test',
        },
        {
            'entity': 'property',
            'type': 'item',
            'value': {
                'entity': 'item',
                'label': 'my_item',
                'property': [
                    # etc.
                ]
            }
        }
    ]
}
```

````python
import pandas as pd
from zipfile import ZipFile

path = "../../data/SNCF/horraires/export-ter-gtfs-horraires-ter.zip"
path = "../../data/Auvergne-Rhone-Alpes/DAT_AURA_GTFS_ExportAOM_Option1.zip"

# declaration of variable used for storing the files from gtfs zip
routes_df = trips_df = calendar_dates_df = stops_df = stop_times_df = None

with ZipFile(path) as myzip:

    myzip.printdir()
    myzip.namelist()

    routes_df = pd.read_csv(myzip.open('routes.txt'), dtype={
        'route_id': 'str',  
        'agency_id': 'str',  
        'route_short_name': 'str',  
        'route_long_name': 'str', 
        'route_desc': 'str', 
        'route_type': 'str',
        'route_url': 'str',
        'route_color': 'str',  
        'route_text_color': 'str',
    })

    trips_df = pd.read_csv(myzip.open('trips.txt'), dtype={
        'route_id': 'str',
        'service_id': 'str',
        'trip_id': 'str',
        'trip_headsign': 'str',
        'direction_id': 'str',
        'block_id': 'str',
        'shape_id': 'str',
    })


    stops_df = pd.read_csv(myzip.open('stops.txt'), dtype={
        'stop_id': 'str',
        'stop_name': 'str',
        'stop_desc': 'str',
        'stop_lat': 'float64',
        'stop_lon': 'float64',
        'zone_id': 'str',
        'stop_url': 'str',
        'location_type': 'str',
        'parent_station': 'str',
    })


    stop_times_df = pd.read_csv(myzip.open('stop_times.txt'), dtype={
        'trip_id': 'str',
        'arrival_time': 'str',
        'departure_time': 'str',
        'stop_id': 'str',
        'stop_sequence': 'int64',
        'stop_headsign': 'str',
        'pickup_type': 'str',
        'drop_off_type': 'str',
        'shape_dist_traveled': 'float64',
    })

    # not used: agency, feed_info
````

````python
i1 = 0
n1 = len(trips_df)

itineraire_dict = [] # liste des dictionnaires représentant les itinéraires

n1 /= 1000 # pour debugg, à supprimer après
while i1 < 1:#n1:
    print('Tour de boucle ', i1, ' sur ', n1)
    itineraire_route_id = trips_df.iloc[i1]['route_id']
    itineraire_trip_id = trips_df.iloc[i1]['trip_id']
    itineraire_direction_id = trips_df.iloc[i1]['direction_id']
    i1+=1

    horraires_arrets_de_l_itineraire = stop_times_df[stop_times_df['trip_id'] == itineraire_trip_id]
    # debugg message, nb arrets in the itineraire
    nb_horraires = len(horraires_arrets_de_l_itineraire)
    if nb_horraires > 25:
        print('Il y a ', nb_horraires, 'arrêts dans cet itinéraire.')


    itineraire_property = []
    
    # Fifth we build the "horaire" entity
    i2 = 0
    nb_horraires = len(horraires_arrets_de_l_itineraire)
    

    # for horraire in horraires_arrets_de_l_itineraire:
    while i2 < nb_horraires:
        horraire = horraires_arrets_de_l_itineraire.iloc[i2]
        horraire_stop_id = horraire['stop_id']
        horraire_trip_id = horraire['trip_id']
        horraire_arrival_time = horraire['arrival_time']

        # Sixth we build the "arret" entity
        arret = stops_df[stops_df['stop_id'] == horraire_stop_id]
        arret_stop_id = arret['stop_id']
        arret_stop_name = arret['stop_name'].values[0]
        arret_stop_lat = arret['stop_lat'].values[0]
        arret_stop_lon = arret['stop_lon'].values[0]
        arret_parent_station = arret['parent_station']

        i2+=1 # impoprtant de laiser l'incrémentation ici pour avoir le bon indice dans le tableau number_in_letter

        # construit la propriété horraire
        horraire_property = {
            'entity': 'property',
            'label': 'horraire '+number_in_letter[i2]+' arret',
            'type': 'string',
            'value': horraire_arrival_time,
        }
        itineraire_property += [horraire_property]

        # construit la propriété arret
        arret_property = {
            'entity': 'property',
            'label': "s arrete en "+number_in_letter[i2],
            'type': 'item',
            'value': {
                'entity': 'item',
                'label': arret_stop_name,
                'property': [
                    {
                        'entity': 'property',
                        'label': 'est un',
                        'type': 'string',
                        'value': 'arret',
                    },
                    {
                        'entity': 'property',
                        'label': 'a pour type arret',
                        'type': 'string',
                        'value': 'arret de train',
                    },
                    {
                        'entity': 'property',
                        'label': 'a pour longitude',
                        'type': 'string',
                        'value': arret_stop_lon,
                    },
                    {
                        'entity': 'property',
                        'label': 'a pour latitude',
                        'type': 'string',
                        'value': arret_stop_lat,
                    },
                ]
            },
        }
        itineraire_property += [arret_property]


    itineraire_property += [{
        'entity': 'property',
        'label': 'est un',
        'type': 'string',
        'value': 'train',
    }]
    # construit la propriété itineraire
    itineraire_dict += [{
        'entity': 'item',
        'label': routes_df[routes_df['route_id'] == itineraire_route_id]['route_long_name'].values[0],
        'property': itineraire_property,
    }]


# lie les itinéraires à la SNCF.
intermediate_rdf_model = {
    'entity': 'item',
    'label': 'SNCF',
    'property': [
        {
            'entity': 'property',
            'label': 'est un',
            'type': 'string',
            'value': 'agence de transports en comun',
        },
        {
            'entity': 'property',
            'label': 'propose',
            'type': 'listItem',
            'value': itineraire_dict,
        }
    ]   
    }


# debugg message, dispklaay the final dict:
pprint(intermediate_rdf_model, width=200)
# print(json.dumps(intermediate_rdf_model, indent=4))
````
