```python
import pandas as pd

root_path = '../../data/Auvergne-Rhone-Alpes/'
def nb_unique_values_per_column(df):
    for col in df.columns:
        print(f"{col}: {df[col].nunique()}")
```

```python
from zipfile import ZipFile
path = root_path+'DAT_AURA_GTFS_ExportAOM_Option1.zip'
with ZipFile(path) as myzip:

    myzip.printdir()
    myzip.namelist()

    routes_df = pd.read_csv(myzip.open('routes.txt'), dtype={
        'route_id': 'str',  
        'agency_id': 'str',  
        'route_short_name': 'str',  
        'route_long_name': 'str', 
        'route_desc': 'str', 
        'route_type': 'Int64',
        'route_url': 'str',
        'route_color': 'str',  
        'route_text_color': 'str',
    })

    trips_df = pd.read_csv(myzip.open('trips.txt'), dtype={
        'route_id': 'str',
        'service_id': 'str',
        'trip_id': 'str',
        'trip_headsign': 'str',
        'direction_id': 'str',
        'block_id': 'int64',
        'shape_id': 'str',
    })

    # not very informative 
    # calendar_dates_df = pd.read_csv(myzip.open('calendar_dates.txt'), dtype={
    #     'service_id': 'str',
    #     'date': 'str',
    #     'exception_type': 'int64',
    # })


    stops_df = pd.read_csv(myzip.open('stops.txt'), dtype={
        'stop_id': 'str',
        'stop_name': 'str',
        'stop_desc': 'str',
        'stop_lat': 'str',
        'stop_lon': 'str',
        'zone_id': 'str',
        'stop_url': 'str',
        'location_type': 'str',
        'parent_station': 'str',
    })


    stop_times_df = pd.read_csv(myzip.open('stop_times.txt'), dtype={
        'trip_id': 'str',
        'arrival_time': 'str',
        'departure_time': 'str',
        'stop_id': 'str',
        'stop_sequence': 'str',
        'stop_headsign': 'str',
        'pickup_type': 'str',
        'drop_off_type': 'str',
        'shape_dist_traveled': 'str',
    })
```
